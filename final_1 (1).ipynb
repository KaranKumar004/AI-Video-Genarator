```python
import subprocess
import threading
import time
import socket

def launch_cloudflared(port=8188):
    """Launch cloudflared tunnel and print public URL"""
    while True:
        time.sleep(0.5)
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        result = sock.connect_ex(('127.0.0.1', port))
        if result == 0:
            break
        sock.close()

    print("\n" + "="*60)
    print("🚀 ComfyUI is ready! Launching cloudflared tunnel...")
    print("="*60 + "\n")

    process = subprocess.Popen(
        ["cloudflared", "tunnel", "--url", f"http://127.0.0.1:{port}"],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE
    )

    for line in process.stderr:
        decoded = line.decode()
        if "trycloudflare.com" in decoded:
            url = decoded[decoded.find("http"):].strip()
            print("🌐 PUBLIC URL:", url)
            print("\n" + "="*60)
            print("📌 COPY THIS URL TO ACCESS COMFYUI")
            print("="*60 + "\n")

# Start tunnel in background
threading.Thread(target=launch_cloudflared, daemon=True, args=(8188,)).start()

# Launch ComfyUI
%cd /content/drive/MyDrive/ComfyUI
!python main.py --listen 0.0.0.0 --port 8188 --preview-method auto
```

```python
# ============================================================================
# COMPLETE READY-TO-RUN SCRIPT - PASTE IN COLAB
# ============================================================================

import requests
import json
import time
import os

COMFYUI_URL = "http://127.0.0.1:8188"

print("🚀 Starting Animation Generation...")
print("=" * 60)

# Complete workflow (no IP-Adapter needed)
workflow = {
  "1": {
    "inputs": {"ckpt_name": "RealVisXL_V4.0.safetensors"},
    "class_type": "CheckpointLoaderSimple"
  },
  "2": {
    "inputs": {
      "model_name": "mm_sdxl_v10_beta.ckpt",
      "beta_schedule": "sqrt_linear (AnimateDiff)"
    },
    "class_type": "ADE_AnimateDiffLoaderGen1"
  },
  "3": {
    "inputs": {
      "motion_model": ["2", 0],
      "model": ["1", 0]
    },
    "class_type": "ADE_ApplyAnimateDiffModel"
  },
  "4": {
    "inputs": {
      "text": "masterpiece, best quality, 8k, ultra detailed, a person walking forward confidently, smooth walking motion, natural gait, professional cinematography, cinematic lighting",
      "clip": ["1", 1]
    },
    "class_type": "CLIPTextEncode"
  },
  "5": {
    "inputs": {
      "text": "static, still, motionless, blurry, low quality, deformed, distorted, ugly, bad anatomy, worst quality, low res, watermark",
      "clip": ["1", 1]
    },
    "class_type": "CLIPTextEncode"
  },
  "6": {
    "inputs": {"width": 1024, "height": 576, "batch_size": 16},
    "class_type": "EmptyLatentImage"
  },
  "7": {
    "inputs": {
      "seed": 42,
      "steps": 25,
      "cfg": 7.5,
      "sampler_name": "euler_ancestral",
      "scheduler": "normal",
      "denoise": 1.0,
      "model": ["3", 0],
      "positive": ["4", 0],
      "negative": ["5", 0],
      "latent_image": ["6", 0]
    },
    "class_type": "KSampler"
  },
  "8": {
    "inputs": {
      "samples": ["7", 0],
      "vae": ["1", 2]
    },
    "class_type": "VAEDecode"
  },
  "9": {
    "inputs": {
      "images": ["8", 0],
      "frame_rate": 8,
      "loop_count": 0,
      "filename_prefix": "walking_animation",
      "format": "video/h264-mp4",
      "pix_fmt": "yuv420p",
      "crf": 19,
      "save_metadata": True,
      "pingpong": False,
      "save_output": True
    },
    "class_type": "VHS_VideoCombine"
  }
}

# Submit to queue
print("📤 Submitting workflow to ComfyUI...")
response = requests.post(
    f"{COMFYUI_URL}/prompt",
    json={"prompt": workflow, "client_id": "colab_automation"}
)

if response.status_code == 200:
    result = response.json()
    prompt_id = result['prompt_id']
    print(f"✅ Queued successfully! Prompt ID: {prompt_id}\n")

    # Monitor progress
    print("⏳ Generating animation (this will take 3-5 minutes)...\n")
    start_time = time.time()

    while True:
        history_response = requests.get(f"{COMFYUI_URL}/history/{prompt_id}")

        if history_response.status_code == 200:
            history = history_response.json()

            if prompt_id in history and 'outputs' in history[prompt_id]:
                elapsed = int(time.time() - start_time)
                print(f"\n✅ Generation complete in {elapsed} seconds!\n")

                # Find video file
                outputs = history[prompt_id]['outputs']
                for node_id, node_output in outputs.items():
                    if 'gifs' in node_output:
                        video_info = node_output['gifs'][0]
                        video_file = video_info['filename']
                        subfolder = video_info.get('subfolder', '')

                        if subfolder:
                            video_path = f"/content/drive/MyDrive/ComfyUI/output/{subfolder}/{video_file}"
                        else:
                            video_path = f"/content/drive/MyDrive/ComfyUI/output/{video_file}"

                        print("=" * 60)
                        print("🎬 YOUR VIDEO IS READY!")
                        print("=" * 60)
                        print(f"📁 Location: {video_path}")
                        print(f"📏 Duration: ~2 seconds (16 frames at 8 fps)")
                        print("=" * 60)

                        # Verify file exists
                        if os.path.exists(video_path):
                            file_size = os.path.getsize(video_path) / (1024 * 1024)
                            print(f"✅ File verified: {file_size:.2f} MB")

                            # Download
                            print("\n📥 Downloading to your computer...")
                            from google.colab import files
                            files.download(video_path)

                        break
                break

        elapsed = int(time.time() - start_time)
        print(f"⏱️  {elapsed}s elapsed...", end='\r')
        time.sleep(5)

else:
    print(f"❌ Error: {response.text}")

print("\n✅ Done! Check your Downloads folder for the video.")
```
